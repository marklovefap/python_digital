```python
import cv2
import numpy as np
from scipy import fftpack as fft
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/baboon.tif',0)
print(img)
print(img.shape)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img,'gray')
```

```python
img_dct = fft.dct( fft.dct(img.T, norm="ortho").T, norm="ortho" )
print(img_dct)
print(img_dct.shape)
print(np.max(img_dct),np.min(img_dct))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img_dct,'gray',vmax=(np.max(img_dct)*0.01),vmin=0)
```

```python
img_idct = fft.idct( fft.idct(img_dct.T, norm="ortho").T, norm="ortho" )
print(img_idct)
plt.figure(figsize=(6,6))
plt.imshow(img_idct,'gray')
```

```python
img_dct2 = img_dct.copy()
img_dct2[10:,10:] = 0
print(img_dct2)
```

```python
img_b0 = img[0:8,0:8].copy()
img_b1 = img[9:17.9:17].copy()
print(img_b0.shape)
print(img_b0)
print(img_b0[0][0])
```

```python
height  = len(img) #one column of image
width = len(img[0]) # one row of image
sliced = [] # new list for 8x8 sliced image 
block = 8
print("The image heigh is " +str(height)+", and image width is "+str(width)+" pixels")
```

```python
#dividing 8x8 parts
currY = 0 #current Y index
for i in range(block,height+1,block):
    currX = 0 #current X index
    for j in range(block,width+1,block):
        sliced.append(img[currY:i,currX:j]-np.ones((8,8))*128) #Extracting 128 from all pixels
        currX = j
    currY = i
    
print("Size of the sliced image: "+str(len(sliced)))
print("Each elemend of sliced list contains a "+ str(sliced[0].shape)+ " element.")
```

```python
print(sliced[0][0][0])
```

```python
DCToutput = []
for part in sliced:
    img_idct = fft.dct( fft.dct(img_dct.T, norm="ortho").T, norm="ortho" )
    DCToutput.append(img_idct)
DCToutput[0][0]
print(len(DCToutput[0][0]))
```
